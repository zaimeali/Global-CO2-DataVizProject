```python
# Import Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Import Dataset
co2_df = pd.read_csv('./co2-global-data.csv', header = 0)
```

```python
data_2020 = co2_df[co2_df.year == 2019]
data_2020 = data_2020[data_2020.iso_code.notna()]
data_2020 = data_2020[data_2020.population.notna()]
data_2020 = data_2020[data_2020.co2.notna()]
data_2020.head()
```

```python

df_grouped=co2_df[co2_df['iso_code'].notna()].groupby(co2_df.country)['co2'].sum()

df_grouped = df_grouped.sort_values(ascending=True)

# Select the top 5 least CO2 producing countries
top_5_least_co2_producing_countries = df_grouped.head(15)

# Print the top 5 least CO2 producing countries
print(top_5_least_co2_producing_countries)
```
